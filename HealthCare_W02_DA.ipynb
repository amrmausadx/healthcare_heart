```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt, numpy as np, seaborn as sns
```

```python
# Download latest version or simply uncomment the heart = 'heart.csv line!!!
path = kagglehub.dataset_download("johnsmith88/heart-disease-dataset")

print("Path to dataset files:", path)
heart = path +r'\heart.csv'
# heart = 'heart.csv';
print(heart)
```

```python
heart = pd.read_csv(heart)
heart.head(100)
```

```python

heart.info()
```

```python
heart.describe()
```

```python
age: Age of the patient
sex: Gender of the patient (1 = male, 0 = female)
cp: Chest pain type (4 types)
trestbps: Resting blood pressure (in mm Hg)
chol: Serum cholesterol in mg/dl
fbs: Fasting blood sugar > 120 mg/dl (1 = true, 0 = false)
restecg: Resting electrocardiographic results (values 0, 1, 2)
thalach: Maximum heart rate achieved
exang: Exercise-induced angina (1 = yes, 0 = no)
oldpeak: ST depression induced by exercise relative to rest
slope: Slope of the peak exercise ST segment
ca: Number of major vessels (0-3) colored by fluoroscopy
thal: Thalassemia (3 = normal, 6 = fixed defect, 7 = reversible defect)
target: Heart disease diagnosis (1 = disease, 0 = no disease)
```

```python
# Set plot style
sns.set(style="whitegrid")
```

```python
# 1. Distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(data=heart, x='age', kde=True, hue='target', multiple="stack")
plt.title('Age Distribution by Heart Disease Diagnosis')
plt.show()
```

<h2 dir='rtl'>هذا الرسم يساعد على تحديد الفئات العمرية الأكثر شيوعًا بين المرضى، والتي يمكن أن تشير إلى أن المرضى في فئة عمرية معينة قد يكونون أكثر عرضة للإصابة بأمراض القلب.
</h2>

```python
# 2. Count plot for chest pain type (cp)
plt.figure(figsize=(10, 6))
sns.countplot(x='cp', data=heart, palette='Set2',hue=heart["target"])
plt.title('Chest Pain Type Distribution')
plt.show()
```

<h2 dir="rtl">هذا الرسم يعرض توزيع المرضى حسب نوع ألم الصدر ويوضح عدد المرضى لكل نوع من أنواع ألم الصدر الأربعة و يمكن أن يساعد في فهم العلاقة بين أنواع معينة من آلام الصدر ومرض القلب على سبيل المثال، قد يكون نوع معين من آلام الصدر مرتبطًا بشكل أكبر مع تشخيص مرض القلب
</h2>

```python
# 3. Boxplot of cholesterol levels by chest pain type
plt.figure(figsize=(10, 6))
sns.boxplot(x='cp', y='chol', data=heart, palette='Set3',hue=heart["target"])
plt.title('Cholesterol Levels by Chest Pain Type')
plt.show()
```

<h2 dir='rtl'>يُستخدم هذا الرسم للمقارنة بين مستويات الكوليسترول لدى المرضى بناءً على أنواع آلام الصدر المختلفة، وقد يكشف عن وجود علاقة بين ارتفاع الكوليسترول وأنواع معينة من آلام الصدر</h2>

```python
# 4. Heatmap of correlations
plt.figure(figsize=(12, 8))
corr_matrix = heart.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

<h2 dir='rtl'>تساعد خريطة الحرارة في تحديد المتغيرات التي تؤثر بشكل أكبر على تشخيص مرض القلب (مثل العمر أو الكوليسترول).</h2>

<h4 dir='rtl'>الارتباطات الإيجابية المهمة:
العمر والمتغيرات الأخرى: إذا كان هناك ارتباط إيجابي بين العمر والمتغيرات مثل الكوليسترول (chol) أو ضغط الدم الراحة (trestbps)، فهذا يشير إلى أن الأفراد الأكبر سنًا قد يكون لديهم مستويات كوليسترول وضغط دم أعلى.
نوع ألم الصدر (cp) وتشخيص مرض القلب (target): إذا كان هناك ارتباط إيجابي بين أنواع معينة من آلام الصدر ومرض القلب، يمكن أن يساعد ذلك في تحديد عوامل الخطر.</h4>
<h4 dir='rtl'>
الارتباطات السلبية المهمة:
معدل نبضات القلب الأقصى (thalach) والعمر: غالبًا ما يكون هذا ارتباطًا سلبيًا مهمًا، مما يعني أنه مع زيادة العمر، ينخفض معدل نبضات القلب الأقصى.
تشخيص مرض القلب (target) وthalach: إذا كان هناك ارتباط سلبي، فهذا يشير إلى أن معدلات نبضات القلب الأقصى المنخفضة ترتبط بمرض القلب.</h4>

```python
# 5. Pairplot of selected features
sns.pairplot(heart[['age', 'trestbps', 'chol', 'thalach', 'target']], hue='target', palette='coolwarm')
plt.show()
```

<h2 dir='rtl'>يساعد هذا الرسم في اكتشاف العلاقات المحتملة بين المتغيرات المختلفة التي قد تساعد في التنبؤ بوجود مرض القلب.</h2>

```python
# 6. Violin plot of oldpeak by heart disease target
plt.figure(figsize=(10, 6))
sns.violinplot(x='target', y='oldpeak', data=heart, palette='muted',hue=heart["target"])
plt.title('Oldpeak Distribution by Heart Disease Diagnosis')
plt.show()
```

<h4 dir='rtl'>Oldpeak هو قياس يستخدم في تخطيط القلب (ECG) يشير إلى مقدار انخفاض الموجة ST بعد الجهد، وغالبًا ما يرتبط بوجود مرض القلب.</h4>
<h4 dir='rtl'>يظهر الرسم كيف يختلف توزيع oldpeak بين المرضى الذين تم تشخيصهم بمرض القلب (target = 1) والذين لا يعانون منه (target = 0).</h4>
<h4 dir='rtl'>ساعد الأطباء والباحثين على فهم العلاقة بين oldpeak ومرض القلب بشكل بصري، مما يمكنهم من تحديد العوامل التي قد تساهم في تشخيص المرض.
ويعكس الرسم كيفية تفاعل العوامل المختلفة مع مرض القلب، مما يدعم التوجه نحو علاجات أفضل واستراتيجيات وقائية.</h4>
